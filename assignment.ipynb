```python
# 1. run a loop from 1 to 400 and print all the odd number using for loop
for i in range(1,401):
  if i%2!=0:
    print(i)
```

```python
# 2. run a loop from 400 to 250 and count the number which are div by 11 and 13
count =0
for i in range(400,249,-1):
  if i%11==0 and i%13==0:
    count +=1
print(count)
```

```python
# 3. run for loop from 60 to 170 and take the total all the number along with the total of number divided by nine.
sum =0
sum_div = 0
for i in range(60,170):
  sum = sum+i
  if i%9==0:
    sum_div += i
print(sum)
print(sum_div)


```
